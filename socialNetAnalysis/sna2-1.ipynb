```python
import snap
import unittest
```

```python
# Creating a graph containing Euler path
def EulerPathGraph():
  graph = snap.TUNGraph.New()
  graph.AddNode(0)
  graph.AddNode(1)
  graph.AddNode(2)
  graph.AddNode(3)
  graph.AddNode(4)
  graph.AddEdge(0,1)
  graph.AddEdge(0,2)
  graph.AddEdge(0,3)
  graph.AddEdge(1,2)
  graph.AddEdge(3,4)
  return graph

# Creating a graph containing Euler circuit
def EulerCircuitGraph():
  graph = snap.TUNGraph.New()
  graph.AddNode(0)
  graph.AddNode(1)
  graph.AddNode(2)
  graph.AddNode(3)
  graph.AddNode(4)
  graph.AddEdge(0,1)
  graph.AddEdge(0,2)
  graph.AddEdge(0,3)
  graph.AddEdge(0,4)
  graph.AddEdge(1,2)
  graph.AddEdge(3,4)
  return graph

def has_euler_path(graph):
  vertices = set()

  # We exclude the zero-degree nodes
  graph.DelZeroDegNodes()

  # We get the degree for each node
  degree = graph.GetDegSeqV()

  '''
  Iterating through the node-degree values
  If a node has odd degree we put the node in the set
  '''
  for i in range(0, len(degree)):
    if degree[i] % 2 == 1 :
      vertices.add(i)

  '''
  If the number of odd-degree nodes are equal to 2
  then the graph has Euler path end returns 'True'
  plus the list of nodes
  '''
  if len(vertices) == 2:
    return True, vertices

  return False, vertices

def has_euler_circuit(graph):
  # We exclude the zero-degree nodes
  graph.DelZeroDegNodes()

  # We get the degree for each node
  degree = graph.GetDegSeqV()

  '''
  Iterating through the node-degree values
  If a node has even degree we increase a counter
  '''
  count = 0
  for i in range(0, len(degree)):
    if degree[i] % 2 ==0 :
      count += 1

  '''
  If the number of even-degree nodes are equal to the number
  of the nodes of the graph after removing the zero-degree
  nodes, then the graph has Euler circuit and return a 'True' value
  otherwise the function returns 'False'
  '''
  if count == graph.GetNodes():
    return True

  return False
```

```python
class TestEulerMethods(unittest.TestCase):

  def test_has_euler_path_but_not_circuit(self):
    # Importing a graph containing Euler path and not circuit
    graph = EulerPathGraph()
    result, vertices = has_euler_path(graph)
    self.assertTrue(result)
    self.assertEqual(len(vertices), 2)

  def test_does_not_have_euler_path(self):
    # Importing a graph not containg Euler path
    graph = EulerCircuitGraph()
    result, vertices = has_euler_path(graph)
    self.assertFalse(result)
    self.assertEqual(len(vertices), 0)

  def test_has_euler_circuit(self):
    # I used circle-graph generator
    graph = snap.GenCircle(snap.TUNGraph, 1000, 10)
    result = has_euler_circuit(graph)
    self.assertTrue(result)
    self.assertTrue(graph.GetNodes()>=1000)

  def test_does_not_have_euler_circuit(self):
    # Importing a graph not containg Euler circuit
    graph = EulerPathGraph()
    result = has_euler_circuit(graph)
    self.assertFalse(result)

```

```python
unittest.main(argv=[''], verbosity=2, exit=False)
```
